```python
# scaping the data 
from langchain.document_loaders import WebBaseLoader
import nest_asyncio

nest_asyncio.apply()

loader = WebBaseLoader(urls)
FASTData = loader.aload()
```

```python
def remove_newlines(document):
    # Remove '\n' characters from the page_content
    cleaned_page_content = document.page_content.replace('\n', '')
    # Update the document object with the cleaned page_content
    document.page_content = cleaned_page_content
    return document

# Apply the cleaning function to each document in the list
cleaned_documents = [remove_newlines(doc) for doc in FASTData]

# Print the cleaned page_content of each document
for doc in cleaned_documents:
    print(doc.page_content)
```

```python
file_name = "FASTData.txt"

# Open the file in write mode
with open(file_name, "w") as file:
    # Loop through each element in the list
    for item in FASTData:
        # Write each element to a new line in the text file
        file.write(f"{item}\n")
```

```python
# Text Splitter
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size=1024,
                                      chunk_overlap=64)


docs = text_splitter.split_documents(FASTData)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings()
```

```python
def store_embeddings(docs, embeddings, sotre_name, path):

    vectorStore = FAISS.from_documents(docs, embeddings)

    with open(f"{path}/faiss_{sotre_name}.pkl", "wb") as f:
        pickle.dump(vectorStore, f)
```
